1. Regression 모델 만들기_data_1.txt

```python
# 1) data_1.txt 파일을 읽어서 데이터를 살펴보고 plot을 그린다.
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data_1.txt', delimiter =',')

X = data[:,0].reshape(-1,1)
y = data[:,1]

plt.figure(figsize=(10,6))
plt.scatter(X[y==0],y[y==0], color="orange", label="False")
plt.scatter(X[y==1],y[y==1], color="blue", label="True")
plt.xlabel("Tumor가 cancer 인 여부 (0 = False, 1 = True)")
plt.ylabel("Tumor size (cm)")
plt.title("Mission1 LogisticRgression")
plt.legend()
plt.grid(True)
```

```python
# 2) 적절한 regression 모델을 생성합시다.

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X,y)
```

```python
# 3) 데이터와 함께 모델이 예측한 선을 plot으로 그립시다.

X_test = np.linspace(-3,5,300).reshape(-1,1)
y_pred = model.predict_proba(X_test)[:,1]

plt.figure(figsize=(10,6))
plt.scatter(X[y==0],y[y==0], color="orange", label="False")
plt.scatter(X[y==1],y[y==1], color="blue", label="True")
plt.plot(X_test,y_pred, color="green", label="Logistic Regression Curve")
plt.xlabel("Tumor가 cancer 인 여부 (0 = False, 1 = True)")
plt.ylabel("Tumor size (cm)")
plt.title("Mission1 LogisticRgression")
plt.legend()
plt.grid(True)
```

```python
# 4) 암일 확률이 50%인 (결정 경계) X의 값은 어떻게 되나요?

coef = model.coef_
intercept = model.intercept_
print(coef, intercept)

X_50 = -intercept / coef
y_50 = model.predict_proba(X_50)[:,0]

print(X_50,y_50)
```

```python
# 5) X=4.5 cm 일 때 암이 malignant 할 확률은 어떻게 되나요?

model.predict_proba([[4.5]])
# 암이 malignant할 확률은 80%
```

2. Regression 모델 만들기_data_2.txt

```python
# 1) data_2.txt 파일을 읽어서 데이터를 살펴보고 plot 을 그립시다.

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data_2.txt',delimiter='\t')

X = data[:,0].reshape(-1,1)
y = data[:,1]

plt.scatter(X,y,label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter plot')
```

```python
# 2) 적절한 regression 모델을 생성합시다.

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
```

```python
# 3) 데이터와 함께 모델이 예측한 선을 plot으로 그립시다.

plt.scatter(X,y,label='Data')
plt.plot(X,y_pred, color='red', label='Linear Regression')
plt.xlabel('약물 치료 횟수')
plt.ylabel('검사 수치')
plt.title('Mission2 Linear Regression')
plt.legend()
```

```python
# 4) 치료를 10회 수행하였을 때 기대되는 검사 수치는 몇이 되나요?

y_pred_10 = model.predict([[10.0]])
print(y_pred_10)
```

```python
#5) 생성한 모델의 기울기, y절편, Mean Squared Error 값은 어떻게 되나요?

from sklearn.metrics import mean_squared_error

print(f"기울기: {model.coef_}")
print(f"절편: {model.intercept_}")

mse = mean_squared_error(y,y_pred)
print(f"MSE: {mse}")
```
