1. AI시대에 개발이 필요한 것인가? AI시대에 개발자가 갖추어야 할 역량은?

: AI는 데이터와 알고리즘에 의존하는 상태이다. AI를 발전시켜 제작하는 것은 개발자가 해야하는 일이다.  
그렇다면 AI를 만들 수 있는 사람들만 개발자의 역할을 한다고 보는가? AI는 문제를 통합적으로 사고하지 못한다. AI를 활용하여 새로운 분야에 적용하는 것은 개발자가 할 일이다. 
AI는 복잡한 일을 단순하게 처리해주는 도구에 지나지 않는다. 이를 다양한 산업 분야에서 얼마나 활용할 수 있는가에 대하여 기본적인 산업에 대한 이해와 AI를 접목시킬 수 있는 능력이 미래 개발자가 갖추어야 할 역량이라고 생각한다.

2. 분류 (classification) 문제에서 이진 분류 (binary classification)와 다중 클래스 분류 (multi-class classification)의 예시를 각각 두 개 들어보세요.


이진 분류(binary classification): 성별예측, 질병유무, 이미지 분류, 스팸메일 분류 등

다중 클래스 분류(multi-class classification): 손글씨 분류, 다양한 질병 분류

3. 와인 데이터에서 알코올 도수(alcohol)와 색상 농도(color_intensity)에 해당하는 데이터에서 scatter plot, historgram 을 생성해보세요.

```python
import pandas as pd
from sklearn.datasets import load_wine

data = load_wine()

pd.DataFrame(data.data, columns=data.feature_names)
```

```python
X = data.data[:,[0,9]]
```

```python
pd.DataFrame(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(100)
y = X* np.random.rand(100)

fig, ax = plt.subplots()
ax.set_title("Scatter plot")
ax.set_xlabel("X")
ax.set_ylabel("y")
ax.scatter(X,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(100)
y = X* np.random.rand(100)

fig, ax = plt.subplots()
ax.set_title("Histogram")
ax.set_xlabel("X")
ax.set_ylabel("y")
ax.hist(y, bins=10)
```
