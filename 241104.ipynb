1. Artificial Neural Network 를 구성하는 층 (layer) 은 어떤 것이 있나요?


: 입력층, 은닉층, 출력층이 있다.
입력층은 입력이 되는 원데이터이며,
은닉층은 입력층과 출력층 사이의 층으로 확률 계산에 필요한 복잡한 결정 경계를 학습하는 부분이다.
출력층은 분류 결과의 확률 값이다.

2. 인공 신경망에서 단층 퍼셉트론이 아닌 여러개의 퍼셉트론으로 구성해야 하는 이유는 무엇인가요?

: 단층 퍼셉트론은 데이터에 따라 결정 경계를 제대로 학습하지 못한다. 단층 퍼셉트론은 선형 분류 문제나 단순한 데이터에서 사용할 수 있다. 퍼셉트론을 여러개 구성하면 비선형 문제, 추상적인 특징 등 복잡하고 고차원적인 데이터를 사용하기에 용이하다.
 인공 신경망은 복잡한 데이터를 학습하는 모델이기 때문에, 여러 층의 퍼셉트론을 사용한다면 데이터를 더 잘 이해하고 처리할 수 있게 된다.

3. 질병의 유무를 분류하는 분류 모델을 ANN으로 구성한다고 가정해 봅시다. 출력층은 몇 개로 구성을 하면 될까요?

: 출력층은 1개가 필요하다.
출력층 1개를 사용하면, 질병에 대한 확률값이 나오기 때문에 그 수치를 사용하여 분류할 수 있다. 예를 들어 0.5 이하면 무, 이상이면 유로 분류할 수 있다.

출력층 2개를 사용하여, 질병이 있는 경우의 확률과 없는 경우의 확률을 각각 받아보는 방법도 있다.
